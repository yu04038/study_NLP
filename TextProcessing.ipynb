## 토큰화 작업 전, 후에는 텍스트 데이터를 용도에 맞게 정제(cleaning) 및 정규화(normalization)

**정제(cleaning) : 갖고 있는 코퍼스로부터 노이즈 데이터를 제거한다.**

**정규화(normalization) : 표현 방법이 다른 단어들을 통합시켜서 같은 단어로 만들어준다.**

USA와 US는 같은 의미를 가지므로, 하나의 단어로 정규화가 가능하다. uh-huh와 uhhuh는 형태는 다르지만 여전히 같은 의미를 갖는다. 이러한 정규화를 거치게 되면, US를 찾아도 USA도 함께 찾을 수 있다.

대소문자 통합의 경우 대문자와 소문자가 구분되어야 하는 경우가 있다. ex) US(나라), us(우리)

## 불필요한 단어의 제거(Removing Unnecessary Words)

정제 작업에서 제거해야하는 노이즈 데이터는 자연어가 아니면서 아무 의미도 갖지 않는 글자들(특수 문자 등)을 의미하기도 하지만, 분석하고자 하는 목적에 맞지 않는 불필요 단어들을 노이즈 데이터라고 하기도 한다.

**(1) 등장 빈도가 적은 단어(Removing Rare words)**

**(2) 길이가 짧은 단어(Removing words with a very short length)** : 영어권 언어에서는 길이가 짧은 단어를 삭제하는 것만으로도 어느정도 자연어 처리에서 크게 의미가 없는 단어들을 제거하는 효과를 볼 수 있다고 알려져 있다.

```python
# 길이가 1~2인 단어들을 정규 표현식을 이용하여 삭제

import re
text = "I was wondering if anyone out there could enlighten me on this car."
shortword = re.compile(r'\W*\b\w{1,2}\b')
print(shortword.sub('', text))
```
